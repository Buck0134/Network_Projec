```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
weekday_data = pd.read_csv('Data/202304-citibike-weekday_data.csv', low_memory=False)
weekday_data['started_at'] = pd.to_datetime(weekday_data['started_at'])
```

```python
num_weekday = 20
num_edges_by_hour_weekday = np.zeros(24)

for i in range(24):
    num_edges_by_hour_weekday[i] = len(weekday_data[weekday_data['started_at'].dt.hour == i])/num_weekday

rush_hours = [7, 8, 9, 16, 17, 18, 19]
color = ['#eb4034' if i in rush_hours else '#1d7ccf' for i in range(24)]
plt.figure(figsize=(8, 4))
plt.bar(np.arange(24), num_edges_by_hour_weekday, color=color)
plt.xticks(np.arange(24), np.arange(24))
plt.ylim(0, 12000)
plt.xlabel('Hour')
plt.title('Number of trips by hour of day (weekday)')

legend_labels = ['Rush Hours', 'Non-Rush Hours']
legend_colors = ['#eb4034', '#1d7ccf']
legend_patches = [plt.Rectangle((0, 0), 1, 1, color=color) for color in legend_colors]
plt.legend(legend_patches, legend_labels, loc='upper left')
plt.show()
```

```python
weekend_data = pd.read_csv('Data/202304-citibike-weekend_data.csv', low_memory=False)
weekend_data['started_at'] = pd.to_datetime(weekend_data['started_at'])
```
